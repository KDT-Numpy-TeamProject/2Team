```python
china_ip = pd.read_csv('중국산업생산량_1990_2023.csv')
```

```python
china_ip
```

```python
yearlist = []
monthlist = []
avglist = []
for i in range(china_ip.shape[0]):
    yearlist.append(china_ip.loc[i,'연도'][:4])
    monthlist.append(china_ip.loc[i,'연도'][-3:-1])
    avglist.append(china_ip.loc[i,'실제'][:-1])
china_ip['년도'] = yearlist
china_ip['월'] = monthlist
china_ip['평균성장율'] = avglist
```

```python
china_ip.drop(columns='연도', inplace=True)
china_ip.drop(columns='실제', inplace=True)
```

```python
china_ip['평균성장율'] = china_ip['평균성장율'].astype('float64')
```

```python
china_ip = china_ip.set_index('년도')
```

```python
meanlist = []
for i in china_ip.index.unique():
    meanlist.append(china_ip.loc[i,'평균성장율'].mean().round(3))

meanlist
```

```python
china_ip
```

```python
end2DF = pd.DataFrame(meanlist, index= np.arange(1990,2024))
```

```python
endDF.head(10)
```

```python
end2DF.rename(columns={0: '평균성장율'}, inplace=True)
```

```python
end2DF = end2DF.rename_axis('년도')
```

```python
end2DF.head(10)
```

```python
endDF.index = endDF.index.astype('int32')
```

```python
finalDF = pd.concat([endDF,end2DF], axis = 1)
```

```python
finalDF.head(10)
```

```python
# 상관계수 파악
finalDF.corr()
```

```python
yeartotal = np.array(finalDF['연합계'])
avg = np.array(finalDF['평균성장율'])

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
# 산점도
ax.scatter(yeartotal, avg, color='green')
ax.set_xlabel('연간 황사 빈도', fontsize='x-large')
ax.set_ylabel('연간 평균 산업성장율', fontsize='x-large')

plt.show()
```

```python
# 계수 B_0과 B_1을 구한다
poly_fit = np.polyfit(yeartotal, avg, 1)
# B_0 + B_1 x를 반환하는 함수를 작성
poly_1d = np.poly1d(poly_fit)
# 직선을 그리기 위해 x좌표를 생성
xs = np.linspace(yeartotal.min(), yeartotal.max())
# xs에 대응하는 y좌표를 구한다
ys = poly_1d(xs)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.plot(xs, ys, color='gray')#, label = f'{poly_fit[1]:.2f}+{poly_fit[0]:.2f}x')
ax.scatter(yeartotal, avg, color='green')#label='score', color='green')
ax.set_xlabel('연간 황사 빈도')
ax.set_ylabel('연간 평균 산업성장율')
# 범례 표시
#ax.legend(loc='upper left')

plt.show()
```

```python
# 히트바
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

c = ax.hist2d(yeartotal, avg, bins=[9,8], range=[(0,25), (2.5,22)])
ax.set_xlabel('연간 황사 빈도')
ax.set_ylabel('연간 평균 산업성장율')
ax.set_xticks(c[1])
ax.set_yticks(c[2])
# 컬러 바 표시
fig.colorbar(c[3], ax=ax)
plt.show()
```
